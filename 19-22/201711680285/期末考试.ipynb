```python
#1.
def the_two ('*',n) :
    line = n * '*'
    for i in range(1,len(line)+1):
        return 2*(n-i)*' ' + '* '*i + n*'*' + ' *'*i

the_two('*',5)
```

```python
#2.
def A(m,n):
    if m == 0:
        A=n+1
    elif m > 0 and n == 0:
        A=A(m-1,1)
    elif m > 0 and n > 0:
        A1=A(m,n-1)
        A=(m-1,A1)
    return A
    
A(3,4)
```
